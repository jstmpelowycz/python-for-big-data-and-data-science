5. (9б). Створіть клас Фігура без атрибутів, з методом get_area для отримання площі фігури, що повертає 0 та __add __, який приймає self та other в якості аргументів, а повертає суму площин фігур self та other.

```python
class Figure:
    def get_area(self):
        return 0

    def __add__(self, other):
        return self.get_area() + other.get_area()
```

6. (11б). Створіть 2 дочірніх класи від Фігури: Трикутник та Коло, які мають атрибути, необхідні для розрахунку площин. Визначте метод get_area в кожному з них так, щоби вони розраховували площу в залежності від формули для кожного типу фігури. Створіть обʼєкт класу Трикутник та обʼєкт класу Коло. Виконайте операцію суми за допомогою оператора + між ними.

```python
import math

class Triangle(Figure):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def get_area(self):
        return (self.base * self.height) / 2


class Circle(Figure):
    def __init__(self, radius):
        self.radius = radius

    def get_area(self):
        return math.pi * self.radius ** 2


triangle = Triangle(base=10, height=5)
circle = Circle(radius=5)

total_area = triangle + circle

print(f"Total area is {total_area}")
```

8. (13б). Створіть клас BankAccount з приватними атрибутами balance та account_number. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class BankAccount:
    def __init__(self, balance, account_id):
        self.__balance = balance
        self.__account_id = account_id

    @property
    def balance(self):
        return self.__balance

    @property
    def account_id(self):
        return self.__account_id

    @balance.setter
    def balance(self, new_balance):
        if new_balance < 0:
            raise ValueError("Incorrect value")

        self.__balance = new_balance

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Incorrect amount")

        self.balance += amount

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Incorrect amount")

        if amount > self.balance:
            raise ValueError("Insufficient costs")

        self.balance -= amount
```

9. (11б). Створіть клас Library, який містить список об'єктів типу Book. Реалізуйте методи для додавання книги, видалення книги та відображення списку книг.

```python
class Book:
    def __init__(self, title, author):
        self.__title = title
        self.__author = author

    def __str__(self):
        return f"«{self.__title}», {self.__author}"


class Library:
    def __init__(self):
        self.__books_set = set()

    def add_book(self, book):
        self.__books_set.add(book)

    def remove_book(self, book):
        self.__books_set.remove(book)

    def show(self):
        [print(book) for book in self.__books_set]
```

10. (13б). Створіть клас Person з атрибутами name та age. Створіть ще один клас Employee з такими атрибутами, як department та salary. Створіть клас Manager, який успадковує обидва класи Person та Employee. Продемонструйте використання множинної спадковості, створивши обʼєкт класу Manager та отримавши mro для цього класу.

```python
class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    def __str__(self):
        return f"Personal info: name is {self.__name}, age is {self.__age}"


class Employee:
    def __init__(self, department, salary):
        self.__department = department
        self.__salary = salary

    def __str__(self):
        return f"Employment info: department is {self.__department}, salary is {self.__salary}"


class Manager(Person, Employee):
    def __init__(self, name, age, department, salary):
        Person.__init__(self, name, age)
        Employee.__init__(self, department, salary)

    def __str__(self):
        personal_info = Person.__str__(self)
        employment_info = Employee.__str__(self)

        return f"{personal_info}\n{employment_info}"


manager = Manager("John Doe", 21, "LMS", 3000)

print(manager)
```

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає літери англійської абетки, поки вони не вичерпаються.

```python
class AlphabetIterator:
    def __init__(self):
        self.__current_letter = ord('a')

    def __iter__(self):
        return self

    def __next__(self):
        if self.__current_letter > ord('z'):
            raise StopIteration

        letter = chr(self.__current_letter)
        self.__current_letter += 1

        return letter


alphabet = AlphabetIterator()

for letter in alphabet:
    print(letter)
```
