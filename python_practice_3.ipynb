3. Створіть рядок зі своїм імʼям латинкою. Напишіть код, який би автоматизовував наступну задачу.

Якщо в імені є літера "а"
 або "i", Виведіть на екран повідомлення "A or I is in the name".



Якщо в імені є літери "o" та "n", Виведіть на екран повідомлення "O and N is in the name".

Також якщо в імені немає літери "о", але є "n", Виведіть на екран повідомлення "Only N is in the name".

В решті випадків виведіть на екран повідомлення "There is no N in the name, but there might be O".

Зверніть увагу, що, наприклад, для імені "Antonina" буде виведено два повідомлення.

```python
def check_name(name):
  if 'a' in name or 'i' in name:
      print("A or I is in the name")

  if 'o' in name and 'n' in name:
      print("O and N is in the name")

  if 'o' not in name and 'n' in name:
      print("Only N is in the name")

  if 'n' not in name:
      print("There is no N in the name, but there might be O")


my_name = "Antonina"

check_name(my_name)
```

4. Створіть список, використовуючи if (не if-else), range та comprehension.

```python
def is_even(value):
  return value % 2 == 0


my_list = [value for value in range(1, 11) if is_even(value)]

print(my_list)
```

5. Створіть словник, використовуючи dict comprehension, що містить if else. Коли слід використовувати comprehensions, а коли їх слід уникати?

```python
map = {
    'a': 100,
    'b': 101,
    'c': 102,
}

next_map = {
    key: value * 2 if value % 2 == 0 else value * 3 + 1
    for key, value in map.items()
}

print(next_map)
```

Я вважаю, що comprehension треба використовувати лише в тих випадках, коли логіка «проста як двері».

Більш комплексну логіку реалізовувати не доцільно, бо це може бути: unreadable, unscalable, barely debuggable.

7. Продемонструйте роботу циклу while. Не можна використовувати True для запуску циклу, щоб уникнути безкінечного циклу!

```python
def binary_search(lst, target):
  i = 0
  j = len(lst) - 1

  while i <= j:
    mid = (i + j) // 2

    if lst[mid] < target:
      i = mid + 1
    elif lst[mid] > target:
      j = mid - 1
    else:
      return mid

  return -1


my_list = list(range(1, 11))

idx = binary_search(my_list, 10)

print(idx)
```

8. Напишіть програму, яка виведе на екран непарні числа в діапазоні від 0 до 20 включно.

```python
def is_odd(value):
  return value % 2 != 0


def program():
  nums = list(range(1, 21))
  odd_nums = list(filter(is_odd, nums))

  print(odd_nums)


program()
```

9. Створіть список з елементами булевого типу або None. Використовуючи цикли, отримайте в результаті список з кортежами, де перший елемент кортежу - індекс(ціле число), а другий елемент - відповідне значення з першого списку.

Наприклад, [True, True, None, False] -> [(0, True), (1, True), (2, None), (3, False)]

```python
bools = [True, True, None, False]

tupled_bools = []

for index, value in enumerate(bools): # using loop, not line comprehension
  tupled_bools.append((index, value))

print(tupled_bools)
```

10. Створіть словник, де ключі - назви книжок, а значення - їхня кількість у наявності в Вашій міні-бібліотеці. Бібліотека має містити щонайменше 6 книжок (6 пар значень у словнику) і щонайбільше 10 (пар значень). Використовуючи цикли, оновіть словник (не створюйте новий) так, щоби кількість книг у наявності збільшилося на 5 кожної книги. Наприклад,
{'It': 3, 'Fault stars': 10, 'Bible': 17, 'Psychological romance': 4, 'Harry Potter': 13} -> {'It': 8, 'Fault stars': 15, 'Bible': 22, 'Psychological romance': 9, 'Harry Potter': 18}

```python
def more_books(lib):
  for key in library:
    library[key] += 5


library = {
    'book-1': 10,
    'book-2': 15,
    'book-3': 20,
    'book-4': 25,
    'book-5': 30,
    'book-6': 35,
}

more_books(library)

print(library)
```

#

12. Гра "Нумо вгадай".
Розробіть просту гру, де користувач має вгадати випадкове число від 1 до 100. Для генерації випадкового числа використовуйте функцію random.randint(a, b) (тут a та b включно). Для отримання числа з консолі використовуйте функцію input(), результат якої обовʼязково явно приведіть до типу int.

Підказка: використовуйте цикл while та умови if-else, щоб повідомити користувачу піказки (напр. "Більше", "Менше") поки користувач не вгадає. По завершенню виведіть кількість спроб, які знадобилися для вгадування.

```python
from random import randint

def game(min, max):
  secret = randint(min, max)

  while True:
    print('Try your guess: ')

    guess = int(input())

    if guess < secret:
      print('Nah, a bit greater')
    elif guess < secret:
      print('Nah, a bit less')
    else:
      print('What a magician!')

      break


game(1, 100)
```
